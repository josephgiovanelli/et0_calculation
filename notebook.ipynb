```python
# naive utils to extract days from timestamp

day = 1
gmt_timezone = 1 

def get_day(hour):
    if hour % 24 == 0:
        global day
        day = day+1
        return day
    else:            
        return day
        
def set_days(df,column_name,new_column):
    df[new_column] = df[column_name].apply(lambda x: get_day((int(x / 3600)+gmt_timezone)%24))
    global day
    day = 1
    return df
```

```python
#daily precipitation & irrigation values

pr = set_days(pr,'timestamp','day').groupby(pr['day'], as_index=False)["precipitation"].sum()
ir = set_days(ir,'timestamp','day').groupby(ir['day'], as_index=False)["irrigation"].sum()
ir['irrigation'] = ir['irrigation'].apply(lambda x: x/2)

et = pd.DataFrame([{'timestamp': 0, 'et0': 0}]).append(et, ignore_index =True)
ir
```

```python
soil_density = 1.3
soil_volume = 30

#Daily avg soil moisture for every sensor
sim_wc = set_days(sim_soil_moisture,'timestamp','day').drop(columns=['timestamp'],axis=1).groupby(sim_soil_moisture['day'], as_index=False).mean()
real_wc = set_days(real_soil_moisture,'timestamp','day').drop(columns=['timestamp'],axis=1).groupby(real_soil_moisture['day'], as_index=False).mean()

#Daily average soil water content on real and simulated data
real_wc = pd.DataFrame(real_wc.mean(axis=1).apply(lambda x: x*soil_density*soil_volume),columns=['SWC']).reset_index(names="day")
sim_wc = pd.DataFrame(sim_wc.mean(axis=1).apply(lambda x: x*soil_density*soil_volume),columns=['SWC']).reset_index(names="day")
```

```python
#Daily water deficit 

#SWC-daily-deficit on real data
real_daily_water_deficit = pd.DataFrame(ir['irrigation'].add(pr['precipitation'].sub(real_wc.diff(axis=0)['SWC'].fillna(0),fill_value=0)),columns=['water deficit'])

#SWC-daily-deficit on simulated data
sim_daily_water_deficit = pd.DataFrame(ir['irrigation'].add(pr['precipitation'].sub(sim_wc.diff(axis=0)['SWC'].fillna(0),fill_value=0)),columns=['water deficit'])

#Et0-daily-water deficit, same for real/simulated data
et_water_deficit = pd.DataFrame(ir['irrigation'].add(pr['precipitation'].sub(et['et0'],fill_value=0)),columns=['water deficit'])
```

```python
#Calculating, for every day, water content of soil an hour before irrigation
def wc_before_one_hour():
    irrigation = pd.read_csv('/home/data/water/irrigation.csv')
    real_soil_moisture = pd.read_csv('/home/data/sim/water_content.csv')

    irrigation['timestamp'] = irrigation['timestamp'].apply(lambda x: datetime.fromtimestamp(x))

    irrigation_hours = []

    for index,row in irrigation.iterrows():
        if index < len(irrigation.index)-1 and row['irrigation'] == 0 and irrigation.iloc[[index+1]]['irrigation'].values[0] > 0 and irrigation.iloc[[index+2]]['irrigation'].values[0] > 0 :
            irrigation_hours.append(row['timestamp'])

    real_soil_moisture['timestamp'] = real_soil_moisture['timestamp'].apply(lambda x: datetime.fromtimestamp(x))

    timestamp =  real_soil_moisture[real_soil_moisture['timestamp'].isin(irrigation_hours)].reset_index(names="day").drop(columns=['day'])
    pre_irrigation_wc = pd.DataFrame(timestamp['timestamp'])
    pre_irrigation_wc['value'] = timestamp.loc[:,[c for c in df.columns if c!= "timestamp"]].mean(axis=1).apply(lambda x: x*soil_density*soil_volume)
    return pre_irrigation_wc

#WC after X hours from irrigation
def wc_after_irrigation(hours):

    irrigation = pd.read_csv('/home/data/water/irrigation.csv')
    real_soil_moisture = pd.read_csv('/home/data/sim/water_content.csv')

    irrigation['timestamp'] = irrigation['timestamp'].apply(lambda x: datetime.fromtimestamp(x))
    
    after_irrigation_hours = []
    for index,row in irrigation.iterrows():
        to_add = False
        for i in range(0,hours):   
            if  index > hours and irrigation.iloc[[index-i]]['irrigation'].values[0] == 0  :
                to_add = True 
            else:
                to_add = False
                break   
        if irrigation.iloc[[index-hours]]['irrigation'].values[0] > 0 and irrigation.iloc[[index-hours-1]]['irrigation'].values[0] > 0 and to_add == True:
            after_irrigation_hours.append(row['timestamp'])

    real_soil_moisture['timestamp'] = real_soil_moisture['timestamp'].apply(lambda x: datetime.fromtimestamp(x))
    timestamp =  real_soil_moisture[real_soil_moisture['timestamp'].isin(after_irrigation_hours)].reset_index(names="day").drop(columns=['day'])

    post_irrigation_wc = pd.DataFrame(timestamp['timestamp'])
    post_irrigation_wc['value'] = timestamp.loc[:,[c for c in df.columns if c!= "timestamp"]].mean(axis=1).apply(lambda x: x*soil_density*soil_volume)
    return post_irrigation_wc
```
