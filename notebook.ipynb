```python
ir = pd.read_csv('/home/data/water/irrigation.csv')
pr = pd.read_csv('/home/data/water/precipitation.csv')
et = pd.read_csv('/home/data/et0/penman_monteith.csv')
sm = pd.read_csv('/home/data/sim/water_content.csv')
```

```python
# naive utils to extract days from timestamp

day = 1
gmt_timezone = 1 

def get_day(hour):
    if hour % 24 == 0:
        global day
        day = day+1
        return day
    else:            
        return day
        
def set_days(df,column_name,new_column):
    df[new_column] = df[column_name].apply(lambda x: get_day((int(x / 3600)+gmt_timezone)%24))
    global day
    day = 1
    return df
```

```python


#Calculating daily precipitation & irrigation values

pr = set_days(pr,'timestamp','day').groupby(pr['day'], as_index=False)["precipitation"].sum()
ir = set_days(ir,'timestamp','day').groupby(ir['day'], as_index=False)["irrigation"].sum()
et = pd.DataFrame([{'timestamp': 0, 'et0': 0}]).append(et, ignore_index =True)

#Daily water deficit: irrigation + precipitation - et0

tw = pd.DataFrame(ir['irrigation'].add(pr['precipitation'].sub(et['et0'],fill_value=0)),columns=['total water'])


tw
```

```python
#Calculating, for every moisture sensor, daily-avg-value.

sm = set_days(sm,'timestamp','day')  
sm = sm.drop(columns=['timestamp'],axis=1)
sm = sm.groupby(sm['day'], as_index=False).mean()

```

```python
soil_density = 1.3
soil_volume = 30

#Daily avg soil moisture by averaging every daily sensor value * soil density * soil volume

sim_wc = pd.DataFrame(sm.mean(axis=1).apply(lambda x: x*soil_density*soil_volume),columns=['SWC']).reset_index(names="day")


sim_wc
```

```python
#Same application, real datas

#Daily avg soil moisture by averaging every daily sensor value * soil density * soil volume

real_wc = pd.read_csv('/home/data/interp_obs/water_content.csv')
real_wc = set_days(real_wc,'timestamp','day')
real_wc = real_wc.drop(columns=['timestamp'],axis=1)
real_wc = real_wc.groupby(sm['day'], as_index=False).mean()

real_wc = pd.DataFrame(real_wc.mean(axis=1).apply(lambda x: x*soil_density*soil_volume),columns=['SWC']).reset_index(names="day")


real_wc
```
