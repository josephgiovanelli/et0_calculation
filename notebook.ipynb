```python
ir = pd.read_csv('/home/data/water/irrigation.csv')
pr = pd.read_csv('/home/data/water/precipitation.csv')
et = pd.read_csv('/home/data/et0/penman_monteith.csv')

sim_soil_moisture = pd.read_csv('/home/data/sim/water_content.csv')
real_soil_moisture = pd.read_csv('/home/data/interp_obs/water_content.csv')
```

```python
# naive utils to extract days from timestamp

day = 1
gmt_timezone = 1 

def get_day(hour):
    if hour % 24 == 0:
        global day
        day = day+1
        return day
    else:            
        return day
        
def set_days(df,column_name,new_column):
    df[new_column] = df[column_name].apply(lambda x: get_day((int(x / 3600)+gmt_timezone)%24))
    global day
    day = 1
    return df
```

```python
#Calculating daily precipitation & irrigation values

pr = set_days(pr,'timestamp','day').groupby(pr['day'], as_index=False)["precipitation"].sum()
ir = set_days(ir,'timestamp','day').groupby(ir['day'], as_index=False)["irrigation"].sum()
et = pd.DataFrame([{'timestamp': 0, 'et0': 0}]).append(et, ignore_index =True)
```

```python
soil_density = 1.3
soil_volume = 30

#Daily avg soil moisture for every sensor
sim_wc = set_days(sim_soil_moisture,'timestamp','day').drop(columns=['timestamp'],axis=1).groupby(sim_soil_moisture['day'], as_index=False).mean()
real_wc = set_days(real_soil_moisture,'timestamp','day').drop(columns=['timestamp'],axis=1).groupby(real_soil_moisture['day'], as_index=False).mean()

#Daily average soil water content on real and simulated data
real_wc = pd.DataFrame(real_wc.mean(axis=1).apply(lambda x: x*soil_density*soil_volume),columns=['SWC']).reset_index(names="day")
sim_wc = pd.DataFrame(sim_wc.mean(axis=1).apply(lambda x: x*soil_density*soil_volume),columns=['SWC']).reset_index(names="day")
```

```python
#Daily water deficit 

#SWC-daily-deficit on real data
real_daily_water_deficit = pd.DataFrame(ir['irrigation'].add(pr['precipitation'].sub(real_wc.diff(axis=0)['SWC'].fillna(0),fill_value=0)),columns=['total water'])

#SWC-daily-deficit on simulated data
sim_daily_water_deficit = pd.DataFrame(ir['irrigation'].add(pr['precipitation'].sub(sim_wc.diff(axis=0)['SWC'].fillna(0),fill_value=0)),columns=['total water'])

#Et0-daily-water deficit, same for real/simulated data
et_water_deficit = pd.DataFrame(ir['irrigation'].add(pr['precipitation'].sub(et['et0'],fill_value=0)),columns=['total water'])
```

```python
# Plotting results

# IN questo caso SWC Ã¨ mooolto piatto nel tempo quindi il delta tende a 0, allora [ir + pr - delta_swc] -> ir , con [pr,delta_swc -> 0]
plt.xticks(range(len(real_wc.index)))
plt.plot(real_wc['day'].index, real_wc['SWC'], label='real-SWC')
plt.plot(real_wc['day'].index, sim_wc['SWC'], label='sim-SWC')
plt.plot(real_wc['day'].index, ir['irrigation'], label='irrigation')
plt.plot(real_wc['day'].index, et_water_deficit['total water'], label='et0-water-deficit')
plt.plot(real_wc['day'].index, sim_daily_water_deficit['total water'], label='sim_SWC-water-deficit')
plt.plot(real_wc['day'].index, real_daily_water_deficit['total water'], label='real_SWC-water-deficit')
plt.plot()
plt.legend(loc='best')
plt.title('Comparison between et0-based daily water deficit & SWC-based')
```

```python
plt.xticks(range(len(real_wc.index)))
plt.plot(real_wc['day'].index, real_wc['SWC'], label='real-SWC')
plt.plot(sim_wc['day'].index, sim_wc['SWC'], label='sim-SWC')
plt.plot()
plt.legend(loc='best')
plt.title('Comparison between real-data/simulated data SWC curve')
```
